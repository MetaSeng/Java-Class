```python
def categories_desc_order(df, categories,values):
  mean = df.groupby(categories)[values].mean()
  desc = mean.sort_values(ascending = False)
  return desc.index
```

```python
#Highes paying counties ( which country offer highest salary)
order = categories_desc_order(df, 'company_location', 'salary_usd')
plt.figure(figsize = (7, 7))
plt.title ("Highest paying counties", fontsize = 14)

sns.barplot(data = df, x='salary_usd', y='company_location', order=order,)
plt.xlabel("Salary (USD)")
plt.ylabel("Country")
```

```python
# Highest paying countries (which country offers the highest salary)
order = categories_desc_order(df, 'company_location', 'salary_usd')

plt.figure(figsize=(7, 7))
plt.title("Highest Paying Countries", fontsize=14)

# Use a colorful palette like 'viridis', 'magma', or 'Set2'
sns.barplot(data=df, x='salary_usd', y='company_location', order=order, palette='inferno')

plt.xlabel("Salary (USD)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

Highest paying jobs

```python
#highest paying jobs
order = categories_desc_order(df, 'job_title', 'salary_usd')
plt.figure(figsize = (7,7))
plt.title("Highest Paying Jobs Ranking", fontsize = 14)

sns.barplot(data = df, x = 'salary_usd', y = 'job_title',order = order,  palette = 'Set2')
plt.xlabel("Salary (USD)")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```
